```python
# download file and save as alice_novel.txt

# open the file and read it into a variable resume
resumetxt = open('C:/Users/dliu/Desktop/Google Drive/2.Resume&CV/Liu-Dunyu-Resume-v3.txt', 'r').read()
    
print ('Resume read and loaded!')
```

```python
stopwords = set(STOPWORDS) # Clear duplicates
```

```python
# instantiate a word cloud object
resume = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
resume.generate(resumetxt)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# display the word cloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(14)

plt.imshow(resume, interpolation='bilinear')
plt.axis('off')
plt.show()
```
